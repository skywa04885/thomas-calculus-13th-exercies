38. # Introduction

    Proof of the limit $\lim_{x \to 3} (3x - 7) = 2$

    Based on this information, we will put some definitions used further in this document bellow:

    $f(x) = (3x - 7),\ L = 2,\ c = 3$

    Now we want to proof the limit by showing that if $x$ get's real close to $c$ that $f(x)$ will get real close to $L$. In this case, the 'how close' we are, will be stored in two variables, in which $\delta$ will be how close $x$ is to $c$, and $\epsilon$ will be how close $f(x)$ is to $L$. Also the following inequalities will be true.

    $0 \lt \epsilon$

    $0 \lt |x - c| \lt \delta$ Delta will be bigger than zero since we're interested in what it approaches, not what it evaluates at.

    $|f(x) - L| \lt \epsilon$

    Now if we substitute

    $0 \lt |x - 3| \lt \delta$ 

    $|(3x - 7) - 2| \lt \epsilon$

    ---

    # Method 1

    Now first, we will find the open interval in which $x$ has to be, in order to have the output error smaller than $\epsilon$.

    $|(3x - 7) - 2| \lt \epsilon$
    
    $|3x - 9| \lt \epsilon$

    $- \epsilon \lt 3x - 9 \lt \epsilon$

    $- \epsilon + 9 \lt 3x \lt \epsilon + 9$
    
    $3 - \frac{\epsilon}{3} \lt x \lt \frac{\epsilon}{3} + 3$

    Which tells us that, if we want to error to be less than $\epsilon$ the input $x$ needs to be in the open interval $(3 - \frac{\epsilon}{3},\  \frac{\epsilon}{3} + 3)$

    Now $\delta$ tells us how near $x$ should be to $c$ and that's why we want to calculate the distance from $c$ to both the endpoints, and then define delta to be the smallest of them. This is because $\delta$ should be in between these values, which aren't always symmetric, and such we need to take the smallest one, which will still satisfy the open interval.

    $\delta = \min{\{ 3 - (3 - \frac{\epsilon}{3}),\ (\frac{\epsilon}{3} + 3) - 3 \}} = \min{\{ \frac{\epsilon}{3},\ \frac{\epsilon}{3} \}}$

    Which tells us that $\delta = \frac{\epsilon}{3}$

    ---

    # Method 2

    This method is more complicated with large polynomials or rationals.

    $|(3x - 7) - 2| \lt \epsilon$
    
    $|3x - 9| \lt \epsilon$

    $|3(x - 3)| \lt \epsilon$

    $3|x - 3| \lt \epsilon$

    $|x - 3| \lt \frac{\epsilon}{3}$

    This shows us that we can set $\delta = \frac{\epsilon}{3}$ and all inequalities will still be true.
